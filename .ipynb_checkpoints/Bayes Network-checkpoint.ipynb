```python

import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from pgmpy.models import BayesianModel
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination
import matplotlib.pyplot as plt
```

```python
import pandas as pd

df = pd.read_csv("heart_disease.csv")
df.head()

```

```python
# Remove duplicates and drop missing values
df = df.drop_duplicates().dropna()
```

```python
df = pd.read_csv("heart_disease.csv")  # Reset the dataframe to original
df = df.drop_duplicates().dropna()

from sklearn.preprocessing import MinMaxScaler

# Normalize numeric columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
scaler = MinMaxScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])

# Discretize continuous features into labeled bins
df['age'] = pd.cut(df['age'], bins=3, labels=['low', 'medium', 'high']).astype(str)
df['chol'] = pd.cut(df['chol'], bins=3, labels=['low', 'medium', 'high']).astype(str)
df['thalach'] = pd.cut(df['thalach'], bins=3, labels=['low', 'medium', 'high']).astype(str)

# Also convert fbs and target to string for consistency
df['fbs'] = df['fbs'].astype(str)
df['target'] = df['target'].astype(str)

from pgmpy.models import BayesianModel
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination

# 1. Define the Bayesian Network structure
model = BayesianModel([
    ('age', 'fbs'),
    ('fbs', 'target'),
    ('target', 'chol'),
    ('target', 'thalach')
])

# 2. Train the model using MLE
model.fit(df, estimator=MaximumLikelihoodEstimator)

# 3. Set up inference engine
infer = VariableElimination(model)

```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plt.hist(df['age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Normalized Age Distribution')
plt.xlabel('Normalized Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.hist(df[df['target'] == 1]['chol'], bins=20, color='salmon', edgecolor='black')
plt.title('Cholesterol Distribution (Heart Disease Present)')
plt.xlabel('Normalized Cholesterol')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(6, 4))
sns.countplot(x='sex', hue='target', data=df)
plt.title('Heart Disease by Gender (0 = Female, 1 = Male)')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Heart Disease')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(df['age'], df['thalach'], alpha=0.7, c=df['target'], cmap='coolwarm', edgecolor='k')
plt.title('Age vs. Max Heart Rate (Thalach)')
plt.xlabel('Normalized Age')
plt.ylabel('Normalized Max Heart Rate (Thalach)')
plt.colorbar(label='Heart Disease (0 = No, 1 = Yes)')
plt.grid(True)
plt.show()
```

```python

# 4. Query 1: P(Heart Disease | age = 'medium')
result1 = infer.query(variables=['target'], evidence={'age': 'medium'})
print("P(Heart Disease | Age = 'medium'):\n", result1)

# 5. Query 2: P(Cholesterol | Heart Disease = '1')
result2 = infer.query(variables=['chol'], evidence={'target': '1.0'})
print("\nCholesterol Distribution | Heart Disease = '1':\n", result2)


```

```python
result1 = infer.query(variables=['target'], evidence={'age': 'medium'})
# Convert result to DataFrame for plotting
df1 = result1.values
labels = result1.state_names['target']

plt.figure(figsize=(5, 4))
plt.bar(labels, df1, color='teal', edgecolor='black')
plt.title('P(Heart Disease | Normalized Age = 0.6)')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Probability')
plt.ylim(0, 1)
plt.grid(True)
plt.show()
```

```python
# Query with correct string evidence
result2 = infer.query(variables=['chol'], evidence={'target': '1.0'})
print("\nCholesterol Distribution | Heart Disease = 1.0:\n", result2)

# Plotting
df2 = result2.values
labels2 = result2.state_names['chol']

plt.figure(figsize=(8, 4))
plt.bar(labels2, df2, color='salmon', edgecolor='black')
plt.title('Cholesterol Distribution | Heart Disease Present')
plt.xlabel('Cholesterol (Discretized States)')
plt.ylabel('Probability')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```
